```python
%matplotlib inline

import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow 

from scipy import misc
import numpy as np
import random
import copy

from gmm_image.gmm_image import predict_pixel_GMM_color_MSE, run_GMM, create_image_from_GMM, rgb2gray

```

### Import and show sample image

```python
rgb_image_arr_nacho = misc.imread('nacho_pic.jpg',mode ='RGB') 

imshow(rgb_image_arr_nacho)
```

```python
rgb_image_arr_starry_night = misc.imread('Starry_Night.jpg',mode ='RGB') 

imshow(rgb_image_arr_starry_night)
```

### Randomly sample RGB colors

```python
total_available_pixels = rgb_image_arr_starry_night.shape[0] * rgb_image_arr_starry_night.shape[1]
total_sampled_pixels = 500

random_values = random.sample(range(0, total_available_pixels), total_sampled_pixels)

rgb_pixel_values_starry_night = rgb_image_arr_starry_night.reshape(total_available_pixels,3)[random_values]
```

### Run the GMM-EM algorithm, show image

```python
gmm, gmm_labels = run_GMM(rgb_pixel_values_starry_night,n_components=100)
create_image_from_GMM(rgb_image_arr_nacho,gmm,show_original = True, gray = False)
```
